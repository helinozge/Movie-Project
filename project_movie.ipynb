```python
#writing the json file 
try :
    with open("data.json","w") as file:
        json_data = json.loads(json.dumps(data))
        print(json_data["results"])
        json.dump(json_data["results"], file, indent=4)
except Exception as e:
    print(f"Failed to write JSON: {e}")    
```

```python
product_columns = ['mug', 't-shirt', 'poster', 'keychain', 'funko', 'gift',
                   'bottle', 'toys', 'figure', 'lego', 'stationery',
                   'costume', 'accessories']
# Create total product count to find top movies
final['total_product_count'] = final[product_columns].sum(axis=1)
# Optional: select top N movies with the most product mentions
top_n = 10
top_movies = final.sort_values(by='total_product_count', ascending=False).head(top_n)
# Reshape the DataFrame for plotting
melted = top_movies.melt(id_vars='movie_name',
                         value_vars=product_columns,
                         var_name='product',
                         value_name='count')
# Plot grouped bar chart
plt.figure(figsize=(14, 7))
sns.set(style='whitegrid')
sns.barplot(data=melted, x='movie_name', y='count', hue='product')
plt.title('Selling Products by Movie')
plt.xlabel('Movie Name')
plt.ylabel('Product Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
