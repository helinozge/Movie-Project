```python
import json
import pandas as pd
import requests
#Api url-address
url = "https://imdb236.p.rapidapi.com/api/imdb/search"

querystring = {"type":"movie","genre":"Animation","averageRatingFrom":"5","rows":"100","startYearFrom":"2024","countriesOfOrigin":"US","sortOrder":"ASC","sortField":"id"}

headers = {
	"x-rapidapi-key": "1e64e6fa40msh3340fd2c8cd7fadp174118jsn30711bc8ecfc",
	"x-rapidapi-host": "imdb236.p.rapidapi.com"
}
#requested to get the file
response = requests.get(url, headers=headers, params=querystring)

print(response.json())
#placed received json file in data 
data = response.json()
```

```python
#writing the json file 
try :
    with open("data.json","w") as file:
        json_data = json.loads(json.dumps(data))
        print(json_data["results"])
        json.dump(json_data["results"], file, indent=4)
except Exception as e:
    print(f"Failed to write JSON: {e}")    
```
